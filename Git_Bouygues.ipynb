```python
import re
from selenium import webdriver
from selenium.common.exceptions import TimeoutException, NoSuchElementException, ElementClickInterceptedException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import chromedriver_autoinstaller
import pandas as pd
import pytz
from datetime import datetime

# Installer chromedriver
chromedriver_autoinstaller.install()

# Configurer les options de Chrome en mode headless
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Initialiser WebDriver
driver = webdriver.Chrome(options=options)

urls = [
    'https://www.bouyguestelecom.fr/telephones-mobiles/apple-iphone-11-noir-64-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/apple-iphone-11-noir-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/apple-iphone-12-noir-64-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/apple-iphone-13-minuit-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/apple-iphone-14-minuit-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/apple-iphone-14-plus-minuit-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/apple-iphone-15-noir-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/apple-iphone-15-plus-noir-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/apple-iphone-15-pro-titane-noir-128-go',
    "https://www.bouyguestelecom.fr/telephones-mobiles/apple-iphone-15-pro-titane-noir-512-go",
    'https://www.bouyguestelecom.fr/telephones-mobiles/apple-iphone-15-pro-max-titane-noir-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/apple-iphone-15-pro-max-titane-noir-512-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/google-pixel-8a-noir-volcanique-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/google-pixel-8-noir-volcanique-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/google-pixel-8-noir-volcanique-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/google-pixel-8-pro-noir-volcanique-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/google-pixel-8-pro-noir-volcanique-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/honor-200-5g',
    'https://www.bouyguestelecom.fr/telephones-mobiles/honor-200-lite-5g',
    'https://www.bouyguestelecom.fr/telephones-mobiles/honor-200-pro-5g',
    'https://www.bouyguestelecom.fr/telephones-mobiles/honor-70-lite',
    'https://www.bouyguestelecom.fr/telephones-mobiles/honor-90-5g-noir-12-go-512-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/honor-90-lite-5g-noir-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/honor-90-smart-noir-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/honor-magic6-lite-5g-noir-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-a05s-noir-64-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-a15-5g-bleu-nuit-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-a15-bleu-nuit-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-a25-5g-bleu-nuit-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-a35-5g-bleu-nuit-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-a55-5g-bleu-nuit-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-a55-5g-bleu-nuit-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-s23-fe-graphite-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-s23-noir-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-s24-noir-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-s24-noir-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-s24-plus-noir-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-s24-ultra-noir-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-s24-ultra-noir-512-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-z-flip5-5g-graphite-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-z-flip5-5g-gris-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-z-flip5-5g-gris-512-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-z-fold5-5g-noir-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-z-fold6-5g-gris-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-z-fold6-5g-gris-512-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/samsung-galaxy-z-fold6-5g-gris-1to',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-13t-noir-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-13t-pro-noir-1-to',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-12-5g',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-12-5g-noir-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-13c-5g',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-13c-noir-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-a3',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-a3-noir-64-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-note-12-4g',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-note-12-4g-bleu-64-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-note-12-5g-noir-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-note-12-pro-5g-noir-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-note-12-pro-plus-5g-noir-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-note-13',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-note-13-5g-noir-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-note-13-noir-128-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-note-13-pro-5g-noir-256-go',
    'https://www.bouyguestelecom.fr/telephones-mobiles/xiaomi-redmi-note-13-pro-plus-5g-noir-256-go',

    ]

data = []

# Fonction pour extraire le texte avec attente explicite et nettoyage du texte
def get_text_by_selector(driver, selector, by=By.CSS_SELECTOR, multiple=False, wait_time=10):
    try:
        if multiple:
            elements = WebDriverWait(driver, wait_time).until(
                EC.presence_of_all_elements_located((by, selector))
            )
            return [element.text.replace('\n', ' ').strip() for element in elements]
        else:
            element = WebDriverWait(driver, wait_time).until(
                EC.presence_of_element_located((by, selector))
            )
            return element.text.replace('\n', ' ').strip()
    except TimeoutException:
        return "TimeoutException"

# Fonction pour dissocier les éléments de financement et nettoyer les données
def dissociate_financing(financing_list):
    financing_data = {
        'Financement Smartphone': '',
        'Facilité de Paiement': '',
        'Comptant': ''
    }

    # Fonction pour nettoyer le texte
    def clean_text(text):
        # Supprimer l'espace entre la virgule et les chiffres
        text = re.sub(r',\s+', ',', text)
        # Remplacer '24 mois' par un placeholder temporaire
        text = text.replace('24 mois', 'PLACEHOLDER')
        # Supprimer les chiffres après la virgule
        text = re.sub(r',\d+', '', text)
        # Garder uniquement les chiffres et les signes plus
        cleaned_text = re.findall(r'\d+|\+', text)
        # Remplacer le placeholder par '24 mois'
        cleaned_text = ' '.join(cleaned_text).replace('PLACEHOLDER', '24 mois')
        return cleaned_text

    for item in financing_list:
        if 'Nouveau Financement smartphone' in item:
            financing_data['Financement Smartphone'] = clean_text(item)
        elif 'Facilité de paiement' in item:
            financing_data['Facilité de Paiement'] = clean_text(item)
        elif 'Comptant' in item:
            financing_data['Comptant'] = clean_text(item)

    return financing_data

# Fonction pour nettoyer la donnée 'odr' afin qu'elle soit uniquement numérique
def clean_odr(odr_text):
    numeric_odr = re.findall(r'\d+', odr_text)
    return int(numeric_odr[0]) if numeric_odr else None

# Liste pour stocker les données extraites
all_data = []

# Parcourir chaque URL pour extraire les données
for index, url in enumerate(urls, start=1):
    driver.get(url)

    # Attendre quelques secondes pour s'assurer que la page est complètement chargée
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.CSS_SELECTOR, "body")))

    # Fermer la fenêtre contextuelle des cookies
    try:
        accept_cookies_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Tout accepter')]"))
        )
        accept_cookies_button.click()
    except TimeoutException:
        print("La fenêtre contextuelle de cookies n'a pas été trouvée.")

    # Fermer d'autres fenêtres contextuelles éventuelles
    try:
        close_popup_button = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, ".elq-actions .elq-dialog-draggable"))
        )
        close_popup_button.click()
    except (TimeoutException, NoSuchElementException):
        pass  # Pas de fenêtre contextuelle trouvée

    # Cliquer sur le bouton pour afficher plus de forfaits
    try:
        show_more_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "#subscription-zone > div > div > a:nth-child(2)"))
        )
        driver.execute_script("arguments[0].click();", show_more_button)
        WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".plan-data-options span")))
    except (TimeoutException, NoSuchElementException, ElementClickInterceptedException):
        print("Le bouton 'Afficher plus' n'a pas été trouvé ou n'a pas pu être cliqué.")

    # Cliquer sur chaque élément .plan-data-options label et extraire les données
    try:
        labels = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".plan-data-options label"))
        )

        for label in labels:
            label_text = label.text
            if "20Go" in label_text or "100Go" in label_text or "Sans forfait" in label_text:
                print(f"Ignoré : {label_text}")
                continue

            label.click()
            data = {}
            data['url'] = url  # Ajouter l'URL scrappée
            data['naming'] = get_text_by_selector(driver, "h1", wait_time=20)
            data['PTA'] = get_text_by_selector(driver, "div:nth-of-type(4) div.column:nth-of-type(1) p.is-marginless.is-level-2", wait_time=20)
            data['Financement'] = get_text_by_selector(driver, ".field label", multiple=True, wait_time=20)
            dissociated_financing = dissociate_financing(data['Financement'])
            data['Comptant'] = dissociated_financing['Comptant']
            raw_odr = get_text_by_selector(driver, "btl-ticket-row:nth-of-type(4) div.is-narrow", wait_time=20)
            data['odr'] = clean_odr(raw_odr)

            # Extraire et imprimer .plan-data-options span après le clic
            spans = driver.find_elements(By.CSS_SELECTOR, ".plan-data-options span")
            for span in spans:
                span_text = span.text
                if "Go" not in span_text and "Sans forfait" not in span_text:  # Filtrer les textes des forfaits
                    data['PTA'] = f"{data['PTA']}, {span_text}" if data['PTA'] else span_text
                    print(span_text)

            # Afficher les données extraites
            print(f"url: {data['url']}")
            print(f"naming: {data['naming']}")
            print(f"PTA: {data['PTA']}")
            print(f"Comptant: {data['Comptant']}")
            print(f"odr: {data['odr']}")

            # Ajouter les données extraites à la liste
            all_data.append(data)
    except TimeoutException:
        print("Les éléments .plan-data-options n'ont pas été trouvés.")

    # Afficher le compteur d'URLs traitées
    print(f"Traitement des URLs : {index}/{len(urls)}")

# Fermer le navigateur
driver.quit()

# Convertir les données en DataFrame pour une meilleure visualisation
df = pd.DataFrame(all_data)

# Supprimer les colonnes "Financement" et celles non souhaitées du DataFrame
df.drop(columns=['Financement', 'Financement Smartphone', 'Facilité de Paiement'], inplace=True, errors='ignore')

# Ajouter la colonne 'Operator' avec la valeur 'Bouygues'
df.insert(0, 'Operator', 'Bouygues')

# Définir le fuseau horaire pour Paris
paris_tz = pytz.timezone('Europe/Paris')

# Obtenir l'heure UTC actuelle et la convertir en heure de Paris
utc_time = datetime.utcnow().replace(tzinfo=pytz.utc)
paris_time = utc_time.astimezone(paris_tz)

# Générer le nom de fichier avec la date et l'heure actuelles
date_str = paris_time.strftime("Bouygues-Bench-%Y-%m-%d_%Hh%M-%S.csv")

# Sauvegarder les données dans un fichier CSV
df.to_csv(date_str, index=False)

print(f"Les données ont été sauvegardées dans le fichier {date_str}")

# Imprimer les données extraites
for data in all_data:
    for key, value in data.items():
        if key not in ['Financement', 'Financement Smartphone', 'Facilité de Paiement']:  # Ne pas imprimer ces clés
            print(f"{key}: {value}")


```
